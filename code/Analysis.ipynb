Create a list of files

```python
import glob
```

```python
path_muster = r'C:\Users\s369577\Desktop\F1 CCTB\data\selected-tiles\selected-tiles'
datei_liste = glob.glob(path_muster + '/out_opt_flow_registered_X10_Y10_*.tif')
endlist = []

for datein in datei_liste:
    ski.io.imread(datein)
    endlist.append(datein)
    #print(endlist)
    #print(datein)
```

Plot a grid of images

```python
image_array = [ski.io.imread(datein) for datein in endlist]
fig, axs = plt.subplots(4, 6, figsize=(25,15))
for i, ax in enumerate(axs.flatten()):
    ax.imshow(image_array[i])
    ax.set_title(f'Image {i+1}')

#plt.show()
```
